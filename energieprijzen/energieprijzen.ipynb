# Run Energy Price Calculation Script

This notebook executes the `bereken_energie_prijzen.py` script to calculate energy prices for Q3 2025 - Q2 2026.

## Setup

```python
# Define parameters
start_date = '2025-07-01'
end_date = '2026-07-01'
supplier = 'GreenChoice'
bron = 'jeroenpuntnl' # Options: 'jeroenpuntnl', 'entsoe'

print(f"Parameters:")
print(f"  Bron: {bron}")
print(f"  Start date: {start_date}")
print(f"  End date: {end_date}")
print(f"  Supplier: {supplier}")
```

## Bereken energieprijzen

```python
import subprocess
import os

# Change to project directory
print(f"Working directory: {os.getcwd()}")

# Build command with parameters
cmd = [
    'python3', 
    f'./bereken_dynamische_stroomprijzen_all_in.py',
    '--bron', bron,
    '--start-date', start_date, 
    '--end-date', end_date,
    '--supplier', supplier
]
print(f"\nRunning command: {' '.join(cmd)}\n")

# Run the bepaal_energie_prijzen.py script
result = subprocess.run(cmd, capture_output=True, text=True)

print(f"STDOUT: {result.stdout}")
if result.stderr:
    print(f"\nSTDERR: {result.stderr}")

print(f"\nReturn code: {result.returncode}")
```
